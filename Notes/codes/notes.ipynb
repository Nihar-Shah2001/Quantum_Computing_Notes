### Quantum Circuit Example

````python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from numpy import pi
 
qreg_q = QuantumRegister(3, 'q')
 
circuit = QuantumCircuit(qreg_q)
 
circuit.barrier(qreg_q[0])
circuit.barrier(qreg_q[1])
circuit.barrier(qreg_q[2])
circuit.h(qreg_q[0])
circuit.h(qreg_q[1])
circuit.id(qreg_q[2])
circuit.barrier(qreg_q[0])
circuit.barrier(qreg_q[1])
circuit.barrier(qreg_q[2])
circuit.rx(pi, qreg_q[0])
circuit.cx(qreg_q[1], qreg_q[2])
circuit.barrier(qreg_q[0])
circuit.barrier(qreg_q[1])
circuit.barrier(qreg_q[2])
circuit.x(qreg_q[0])
circuit.id(qreg_q[1])
circuit.z(qreg_q[2])
circuit.barrier(qreg_q[0])
circuit.barrier(qreg_q[1])
circuit.barrier(qreg_q[2])
# @columns [0,0,0,1,1,2,2,2,3,3,4,4,4,5,5,6,6,6]
 
circuit.draw(output='mpl')#.savefig('../images/qcircuit1.png')
````

### Teleportation Circuit

````python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from numpy import pi
 
qreg_q = QuantumRegister(2, 'q')
creg_c = ClassicalRegister(2, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)
 
circuit.cx(qreg_q[0], qreg_q[1])
circuit.h(qreg_q[0])
circuit.barrier()
circuit.measure(qreg_q[0], creg_c[0])
circuit.measure(qreg_q[1], creg_c[1])
# @columns [0,1,2,3]
 
circuit.draw(output='mpl')#.savefig('../images/teleportation2.png')
````

````python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from numpy import pi
 
qreg_q = QuantumRegister(3, 'q')
creg_cx = ClassicalRegister(1, 'c')
creg_cz = ClassicalRegister(1, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)
 
circuit.h(qreg_q[1])
circuit.cx(qreg_q[1], qreg_q[2])
circuit.barrier(qreg_q[1])
circuit.barrier(qreg_q[2])
circuit.cx(qreg_q[0], qreg_q[1])
circuit.h(qreg_q[0])
circuit.barrier()
circuit.measure(qreg_q[0], creg_c[0])
circuit.measure(qreg_q[1], creg_c[1])
circuit.x(qreg_q[2]).c_if(creg_c[1], 1)
circuit.z(qreg_q[2]).c_if(creg_c[0], 1)
# @columns [0,1,2,2,3,4,4,5,6,6,7,8]
 
circuit.draw(output='mpl')#.savefig('../images/teleportation.png')
````

### Superedense Coding

````python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from numpy import pi
 
qreg_q = QuantumRegister(2, 'q')
 
circuit = QuantumCircuit(qreg_q)
 
circuit.h(qreg_q[0])
circuit.cx(qreg_q[0], qreg_q[1])
# @columns [0,1]
 
circuit.draw(output='mpl')#.savefig('../images/superdense-coding1.png')
````

````python
# Create a Quantum Circuit
qubits=QuantumRegister(2,name="q")
cbits=ClassicalRegister(2,name="c")
circuit = QuantumCircuit(qubits,cbits)

q0,q1=qubits
c0,c1=cbits

# Prepare the state to be teleported
circuit.h(q0)
circuit.cx(q0, q1)

circuit.barrier()

oracle_qc=QuantumRegister(2)
oracle=QuantumCircuit(oracle_qc,name='Oracle')

circuit.append(oracle, [q1,q0])

circuit.barrier()

circuit.cx(q0, q1)
circuit.h(q0)

circuit.barrier()

circuit.measure(q0, c0)
circuit.measure(q1, c1)


circuit.draw(output='mpl')#.savefig('../images/superdense-coding4.png')
````

### Deutsch Algorithm

````python
from qiskit import QuantumCircuit
qasm = """
OPENQASM 2.0;
include "qelib1.inc";
gate oracle a, b {
  barrier a;
  barrier b;
  cx a, b;
  barrier a;
  barrier b;
}

qreg q[2];
creg c[1];

h q[0];
x q[1];
h q[1];
oracle q[0], q[1];
h q[0];
measure q[0] -> c[0];

// @columns [0,0,1,2,3,4]

"""
circuit = QuantumCircuit.from_qasm_str(qasm)

circuit.draw(output='mpl')#.savefig('../images/deutsch.png')
````

### Deutsch-Jozsa Algorithm

````python
from qiskit import QuantumCircuit
qasm = """
OPENQASM 2.0;
include "qelib1.inc";
gate oracle a, b, c {
  CX a, c;
  cx b, c;
}
qreg q[3];
creg c[2];

h q[0];
h q[1];
h q[2];
oracle q[0], q[1], q[2];
h q[0];
h q[1];
measure q[0] -> c[0];
measure q[1] -> c[0];

// @columns [0,0,0,0,1,2,2,3,4]

"""
circuit = QuantumCircuit.from_qasm_str(qasm)

circuit.draw(output='mpl')#.savefig('../images/deutsch-jozsa-algorithm.png')
````
